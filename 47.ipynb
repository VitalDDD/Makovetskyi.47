```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sa
from statsmodels.tsa.seasonal import seasonal_decompose

series = pd.read_csv('AirPassengers.csv')
series.head()
```

```python
series.isnull().sum()
```

```python
series = series.rename({'#Passengers' : 'Passengers'}, axis = 'columns') 
series.index = pd.to_datetime(series['Month'])
del series['Month']
series.index.freq = series.index.inferred_freq

matplotlib.rcParams['figure.figsize'] = [10, 5]

series.plot()
pyplot.title("Графік часового ряду", fontsize=13)
pyplot.show()
```

```python
acf_value = acf(series.Passengers)
pacf_value = pacf(series.Passengers, nlags = 3)
acf_value
```

```python
pacf_value
```

```python
matplotlib.rcParams['figure.figsize'] = [10, 3]
plot_acf(series.Passengers)
pyplot.show()
```

```python
plot_pacf(series.Passengers, lags =3, method='ywm')
pyplot.show()
```

```python
print(series.Passengers)
```

```python
data_train = series.Passengers[:115]
decompose = seasonal_decompose(series.Passengers, model='additive', period=12)
decompose_train = seasonal_decompose(data_train, model='additive', period=12)
data_decompose = decompose_train.resid
model_1 = sa.tsa.ARIMA(data_decompose, order=(12, 2, 0)).fit()
```

```python
model_1.summary()
```

```python
pred_1 = model_1.predict(115,143)
```

```python
decompose.resid.plot()
pred_1.plot(style='r--')
```

```python
comp = series.Passengers[114:143]
```

```python

mae = mean_absolute_error(comp, pred_1)
mse = mean_squared_error(comp, pred_1)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}') 
```

```python
model_2 = sa.tsa.statespace.SARIMAX(data_train, 
                                  seasonal_order=(2,2,2,12)).fit()
```

```python
pred_2 = model_2.predict(115,143, typ='levels')
series.plot()
pred_2.plot(style='r--')
```

```python

mae = mean_absolute_error(comp, pred_2)
mse = mean_squared_error(comp, pred_2)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error: {mae:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}') 
```
